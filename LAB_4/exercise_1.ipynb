```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Zadanie 2
Przygotowanie danych do interpolacji. Wczytaj dane z pliku CSV lub innych źródeł danych. Zidentyfikuj i obsłuż brakujące wartości danych. Przeprowadź wstępną analizę danych w celu określenia odpowiedniej metody interpolacji.

```python
# Wczytanie danych
data = pd.read_csv('data/temperatures.csv')
# Zamiana times na date bez potrzebnych godzin
data['time'] = pd.to_datetime(data['time']).dt.date
# Usunięcie rekordów gdzie 'time' się powtarza
data = data.drop_duplicates(subset='time')
# Sprawdzenie Nowych wymiarów
data.shape
data.head()
```

```python
# Wyświetlenie wykresu temperatury
plt.plot(data['time'], data['temperature'])
# Wyświetlanie tylko część legendy na osi x
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.xticks(data['time'][::100], rotation=45)
```

```python
# Tworzenie "Dziur" w danych abyśmy mogli zastosować interpolacje
# Stwórz Listę 5 kopii danych z różnymi dziurami
temp = []
n = 5
for a in range(n):
    temp.append(data.copy())
for a in range(n):
    temp[a]['temperature'] = temp[a]['temperature'].mask(np.random.random(data['temperature'].shape) < 0.8)

plt.scatter(data['time'], temp[0]['temperature'])
```

# Interpolacja liniową aby uzupełnić brakujące dane

```python
# Wymiary wykresów
plt.rcParams['figure.figsize'] = [15, 15]
# Funkcja interpolacji liniowej
temp[0]['temperature'] = temp[0]['temperature'].interpolate(method='linear')
# Wizualizacja danych po interpolacji
plt.plot(temp[0]['time'], temp[0]['temperature'], color='red')
plt.scatter(data['time'], data['temperature'], color='green')
```

# Interpolacja Wielomianowa aby uzupełnić brakujące dane

```python
# Wymiary wykresów
plt.rcParams['figure.figsize'] = [15, 15]
# Funkcja interpolacji liniowej
temp[1]['temperature'] = temp[1]['temperature'].interpolate(method='polynomial', order=3)
# Wizualizacja danych po interpolacji
plt.plot(temp[1]['time'], temp[1]['temperature'], color='red')
plt.plot(data['time'], data['temperature'], color='yellow')
# ustaw grubość lini zielonej na 0.5
plt.setp(plt.gca().get_lines()[0], linewidth=3)
plt.setp(plt.gca().get_lines()[1], linewidth=0.6)
```
