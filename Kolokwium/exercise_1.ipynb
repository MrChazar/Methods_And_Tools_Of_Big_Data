```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# wczytywanie i dzielenie
digits = load_digits()
x = digits.data
y = digits.target
print(f"X: {x}")
print(f"Y: {y}")
```

```python
#scaler = StandardScaler()
#x = scaler.fit_transform(x)
```

```python
# Stosowanie PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)
# Wy≈õwietlenie nowych danych po zastosowaniu PCA
print("Nowe dane po zastosowaniu PCA:")
print(x_pca)
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y, cmap='hsv', edgecolor='k', alpha=0.7)
plt.title('Wyniki PCA dla datasetu Digits')
plt.xlabel('Sk≈Çadowa g≈Ç√≥wna 1')
plt.ylabel('Sk≈Çadowa g≈Ç√≥wna 2')
plt.grid(True)
# Dodanie legendy z etykietami klas
plt.colorbar(scatter, label='Target')
plt.show()
```

```python
# PCA s≈Çabo siƒô sprawia lepszy jest LDA
# Stosowanie TSNE
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(x)
# Wy≈õwietlenie nowych danych po zastosowaniu TSNE
print("Nowe dane po zastosowaniu TSNE:")
print(X_tsne)
```

# Wnioski
PCA s≈Çabo siƒô sprawia w tym zadaniu z powodu zastosowania techniki liniowej

```python
# Wykres wynik√≥w TSNE
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='hsv', edgecolor='k', alpha=0.7)
plt.title('Wyniki analizy g≈Ç√≥wnych sk≈Çadowych (TSNE) dla datasetu Digits')
plt.xlabel('Sk≈Çadowa g≈Ç√≥wna 1')
plt.ylabel('Sk≈Çadowa g≈Ç√≥wna 2')
plt.grid(True)
# Dodanie legendy z etykietami klas
plt.colorbar(scatter, label='Target')
plt.show()
```
